```python
def optimize_SARIMA(endog: Union[pd.Series, list], order_list: list, d: int, D: int, s: int) -> pd.DataFrame:
    
    results = []
    
    for order in tqdm_notebook(order_list):
        try: 
            model = SARIMAX(
                endog, 
                order=(order[0], d, order[1]),
                seasonal_order=(order[2], D, order[3], s),
                simple_differencing=False).fit(disp=False)
        except:
            continue
            
        aic = model.aic
        results.append([order, aic])
        
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q,P,Q)', 'AIC']
    
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC', ascending=True).reset_index(drop=True)
    
    return result_df
```

```python
# Import SARIMA model
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define SARIMA model parameters
order = (2, 1, 1)  # Specify the SARIMA order (p, d, q)
seasonal_order = (2, 0, 0, 12)  # Specify the seasonal order (P, D, Q, m)

# Fit SARIMA model
sarima_model = SARIMAX(df['Price'], order=order, seasonal_order=seasonal_order)
sarima_result = sarima_model.fit()

# Predict 24 months ahead with SARIMA
sarima_forecast = sarima_result.forecast(steps=24)

# Naive method
last_price = df['Price'].iloc[-1]  # Get the last observed price
naive_forecast = [last_price] * 24  # Naive forecast assumes the last price continues for the next 24 months


# Create a DataFrame for the forecasts
forecast_df = pd.DataFrame({
    'SARIMA Forecast': sarima_forecast,
    'Naive Forecast': naive_forecast
})

# Save the DataFrame to an Excel file
forecast_df.to_excel('C:/Users/Dell/Desktop/IIT_KGP/Sem 2/Project 1/New folder/Images/next 24 months forecasts_duplicate.xlsx')

# Display forecasts
print("SARIMA forecast:", sarima_forecast)
print("Naive forecast:", naive_forecast)
```
